# REMOVING $ SIGN FROM COLUMN MONTHLY REVENUE

```python
df["Monthly Revenue"] = df["Monthly Revenue"].str.replace("$","",regex= True)
df
```

```python
df["Monthly Revenue"] = df["Monthly Revenue"].str.replace(".00","",regex= True)
df
```

# CHECKING FOR NULL VALUES

```python
df.isnull().sum()
```

# DROPING ROWS WITH NULL VALUES

```python
df.dropna(axis =0, inplace= True)
df.isnull().sum()
```

# REMOVING UNWANTED HEADING AND TRAILING SPACE FROM COLUMN MONTHLY REVENUE

```python
df["Monthly Revenue"] = df["Monthly Revenue"].str.strip()
df
```

# REMOVING ? SIGN FROM COLUMN USER ID

```python
df["User ID"] = df["User ID"].str.replace("?","",regex= True)
df
```

# DROPING UNWANTED COLUMN "DATA" 

```python
df.drop("Data",axis=1,inplace= True)
df
```

# REMOVING "MONTH" FROM COLUMN PLAN DURATION

```python
df["Plan Duration"] = df["Plan Duration"].str.replace("Month","",regex= True)
df
```

# DEALING WITH CATEGOCICAL DATA

```python
df["Gender"]=df["Gender"].map({"Male":1,"Female":0})  #HERE CONVERTING MALE AS 1 AND FEMALE AS 0
df
```

# CHECKING FOR DUPLICATE RECORDS

```python
df.duplicated().sum()
```

# DROPING DUPLICATE ROWS

```python
df.drop_duplicates(inplace = True)
df.duplicated().sum()
```

# RENAMING THE COLUMN LAST PAYMENT DATE AS PAYMENT DATE

```python
df.rename(columns = {"Last Payment Date":"Payment date"},inplace = True)
df
```

# CREATING NEW COLUMN AS MONTH FROM COLUMN JOIN DATE

```python
df["Month"] = df["Join Date"].dt.month_name()
df
```

# CHANGING THE DATA TYPE OF COLUMN PLAN DURATION TO "INT"

```python
df["Plan Duration"] = df["Plan Duration"].astype("int")
df.dtypes
```

# HOT INCODING OF COLUMN DEVICE

```python
df= pd.get_dummies(df,columns=["Device"])
df
```
